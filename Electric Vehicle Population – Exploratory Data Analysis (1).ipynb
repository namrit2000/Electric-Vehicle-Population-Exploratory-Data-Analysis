```python
ms_values = ["n.a.", "NA", "n/a", "na", 0]
ES = pd.read_csv('Electric_Vehicle_Population_Data.csv', na_values = ms_values)
ES
```

```python
ES.dropna(axis = 0, inplace = True)
ES.isna().sum()
```

```python
ES
```

```python
ES['Postal Code'] = ES['Postal Code'].astype(str)
ES['Model Year'] = ES['Model Year'].astype(str)
ES['DOL Vehicle ID'] = ES['DOL Vehicle ID'].astype(str)
ES.info()
```

```python
#3
ES['Electric Vehicle Type'].value_counts().plot(kind='bar')
plt.title('Dist of EV Types')
plt.xlabel('Vcle Type')
plt.ylabel('Cnt')
plt.xticks(rotation=0)
plt.show()
```

```python
tmake = ES['Make'].value_counts().head(5)
tmake.plot(kind='bar')
plt.title('Top 5 Vehicle')
plt.xlabel('Make')
plt.ylabel('Cnt')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.hist(ES['Electric Range'], bins=20, edgecolor='k')
plt.title('ERange Dist')
plt.xlabel('ERange')
plt.ylabel('Cnt')
plt.show()
```

```python
ES['County'].value_counts().plot(kind='bar')
plt.title('CW Dist of EV')
plt.xlabel('County')
plt.ylabel('No. of EVS')
plt.xticks(rotation=90)
plt.show()
```

```python
#4
plt.scatter(ES['Base MSRP'], ES['Electric Range'], alpha=0.5)
plt.title('Prc vs. ERange')
plt.xlabel('Base MSRP')
plt.ylabel('ERange')
plt.xticks(rotation=90)
plt.show()
```

```python
ec = ES['Clean Alternative Fuel Vehicle (CAFV) Eligibility']

ecnts = ec.value_counts()
tcnt = len(ec)


plt.pie(ecnts, labels=ecnts.index, autopct='%1.1f%%', startangle=140, colors=['yellow', 'red'])
plt.axis('equal')

plt.title('Proportion of Eligible vs. Not Eligible Vehicles')
plt.tight_layout()
plt.show()
```
